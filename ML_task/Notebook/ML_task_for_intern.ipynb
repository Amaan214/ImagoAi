```python
# Checking for outliers by visualizing the dataset
import plotly.express as px
fig = px.bar(original_dataset, y = 'vomitoxin_ppb', title = 'Checking vomitoxin_ppb levels for outliers')
fig.show()
```

**Using InterQuartile Range for extreme values.**

```python
Q1 = original_dataset["vomitoxin_ppb"].quantile(0.25) # 1st quartile is 25%
Q3 = original_dataset["vomitoxin_ppb"].quantile(0.75) # 2nd quartile is 75%

IQR = Q3 - Q1 # calculating interquartile range

# Defining outliers bound
Lower_bound = Q1 - 1.5 * IQR
Upper_bound = Q3 + 1.5 * IQR

# Counting outliers
outliers = original_dataset[((original_dataset["vomitoxin_ppb"] < Lower_bound) | (original_dataset["vomitoxin_ppb"] > Upper_bound))]
"""
-> In Pandas we cannot use if statement therefore we use the condition directly on the dataset to get the outliers.
-> We use  |(Bitwise OR) operation since we cannot use Logical OR condition on elements in pandas.
"""

print(f"Lower bound: {Lower_bound}")
print(f"Upper bound: {Upper_bound}")
print(f"Total number of outliers: {outliers.shape[0]}")
print("")
print(f"Number of outliers: {outliers}")
```

### **Compairing different methods for outliers handling**

```python
# Median handling
median_values = original_dataset["vomitoxin_ppb"].median()
original_dataset["modified_vomitoxin_ppb"] = original_dataset["vomitoxin_ppb"].apply(lambda x : median_values if x < Lower_bound or x > Upper_bound else x)
```

```python
# Log transformation handling
original_dataset["log_vomitoxin_ppb"] = np.log1p(original_dataset["vomitoxin_ppb"])
```

```python
# Winsorsizing handling
lower_threshhold = original_dataset["vomitoxin_ppb"].quantile(0.05)
upper_threshold = original_dataset["vomitoxin_ppb"].quantile(0.95)

winsorsized_data = []
for x in original_dataset["vomitoxin_ppb"]:
  if x < lower_threshhold:
    winsorsized_data.append(lower_threshhold)
  elif x > upper_threshold:
    winsorsized_data.append(upper_threshold)
  else:
    winsorsized_data.append(x)

original_dataset["winsorsized_vomitoxin_ppb"] = winsorsized_data
```

```python
# ----------visualizing-------------------------
fig, axes = plt.subplots(2,2, figsize=(12, 10))
plt.suptitle("Compairing different methods for outliers handling")

# Original
sns.histplot(original_dataset['vomitoxin_ppb'], ax = axes[0,0], kde=True, color = "red")
axes[0,0].set_title('Original Handling')

# median
sns.histplot(original_dataset['modified_vomitoxin_ppb'], ax = axes[0,1], kde=True, color = "blue")
axes[0,1].set_title('Median Handling')

# Log transformed Handling
sns.histplot(original_dataset['log_vomitoxin_ppb'],  ax = axes[1,0], kde = True, color = "green")
axes[1,0].set_title('Log Tranformation Handling')

# Winsorsized
sns.histplot(original_dataset['winsorsized_vomitoxin_ppb'], ax = axes[1,1], kde=True, color = "purple")
axes[1,1].set_title('Winsorsized Handling')


print("\nDescriptive Statistics Comparison:")
print(original_dataset[['modified_vomitoxin_ppb', 'log_vomitoxin_ppb', 'winsorsized_vomitoxin_ppb']].describe())
```

```python
# Normalizing the outliers
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()
original_dataset.iloc[:, 1:] = scaler.fit_transform(original_dataset.iloc[:, 1:])
```

**Visualizing Spectral Bands**

# **2. Dimensionality Reduction**

```python
# Implementing PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
principal_component = pca.fit_transform(original_dataset[spectral_columns])

percentage = np.cumsum(pca.explained_variance_ratio_)
print(f"Explained Variance: {percentage* 100}")
pca.components_
```

```python
pca_df = pd.DataFrame(data = principal_component, columns=['PCA1', 'PCA2', 'PCA3'])

fig = px.scatter_3d(pca_df, x="PCA1", y = "PCA2", z = "PCA3", color=original_dataset['vomitoxin_ppb'], title="3D Scatter Plot(PCA1 vs PCA2 VS PCA3)",labels={"PC1": "Principal Component 1","PC2": "Principal Component 2", "PC3": "Principal Component 3"}, color_continuous_scale="Viridis")
fig.show()
```

# **3. Model Training**

```python
from sklearn.model_selection import train_test_split

x = original_dataset[spectral_columns]
y = original_dataset['vomitoxin_ppb']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=40)
print("Test data size: ", x_test.shape[0])
print("Train data size: ", x_train.shape[0])
```

### **Training CNN Model on Training set**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Flatten, Dense, Dropout, MaxPooling1D

# Define CNN model
cnn_model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation="relu", input_shape=(x_train.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=32, kernel_size=3, activation="relu"),
    Flatten(),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(1, activation="linear")  # Regression output
])

cnn_model.compile(optimizer="adam", loss="mse", metrics=["mae"])
cnn_model.summary()

# Train CNN model
cnn_history = cnn_model.fit(x_train, y_train, epochs=50, batch_size=16, validation_data=(x_test, y_test))
```

# **4. Model Evaluation**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make predictions
y_pred = cnn_model.predict(x_test)

# Compute regression metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# Print evaluation metrics
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```
